```python
img2=cv2.imread(r"\Users\Rahul\Pictures\Screenshots\Screenshot 2023-10-31 125831.png")
(height ,width)=img2.shape[0:2]
zoom=cv2.resize(img2,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)

```

```python
zoom=cv2.resize(img2,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
#rotate image 30'
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img2,m,(height,width))
plt.imshow(rotate)
```

```python
#translate
import numpy as np
n=np.float32([[1,0,100],[0,1,200]])
dst=cv2.warpAffine(img2,n,(height,width))
plt.imshow(dst)
```
